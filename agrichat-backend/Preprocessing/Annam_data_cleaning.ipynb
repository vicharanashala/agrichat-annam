```
import pandas as pd
import numpy as np

# Load the data
df = pd.read_csv("data.csv", low_memory=False)

# Step 1: Identify rows where 'Sector' is numeric
def is_numeric(x):
    try:
        float(x)
        return True
    except:
        return False

# Step 2: Filter rows where Sector is numeric
numeric_sector_df = df[df['Sector'].apply(is_numeric)].copy()

# Step 3: Replace numeric values in Sector with NaN
numeric_sector_df['Sector'] = np.nan

# Step 4: Save the filtered and modified data to a new CSV
numeric_sector_df.to_csv("numeric_sector_cleaned.csv", index=False)

print("✅ Done! File 'numeric_sector_cleaned.csv' saved.")
```

```
import pandas as pd

# Load original file
df = pd.read_csv("data.csv", low_memory=False)

# Function to check if a value is numeric
def is_numeric(value):
    try:
        float(value)
        return True
    except:
        return False

# Remove rows where 'Sector' is numeric (modify in-place)
df = df[~df['Sector'].apply(is_numeric)]

# Overwrite the original file
df.to_csv("data.csv", index=False)

print("Original file updated: Numeric 'Sector' rows removed.")
```

```
import pandas as pd
import numpy as np

class DataCleaner:
    def __init__(self, file_path):
        """
        Initialize with a file path (CSV).
        """
        self.file_path = file_path
        self.df = pd.read_csv(file_path, low_memory=False)

    def remove_columns(self, columns_to_remove):
        """
        Remove given columns from the DataFrame.
        """
        self.df.drop(columns=columns_to_remove, inplace=True, errors='ignore')

    def save_numeric_rows(self, column_name, output_csv):
        """
        Save rows with numeric values in the specified column to a separate CSV,
        and replace those values with NaN in the original DataFrame.
        """
        # Create a mask for numeric values (ignoring NaN)
        mask = pd.to_numeric(self.df[column_name], errors='coerce').notna()

        # Save numeric rows to new file
        numeric_df = self.df[mask]
        numeric_df.to_csv(output_csv, index=False)

        # Replace numeric entries with NaN in the original DataFrame
        self.df.loc[mask, column_name] = np.nan

    def save_cleaned_file(self, output_csv):
        """
        Save the modified DataFrame to a new file.
        """
        self.df.to_csv(output_csv, index=False)
```
