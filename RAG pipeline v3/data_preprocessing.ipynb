```python
def is_numeric_or_punctuation(series: pd.Series) -> pd.Series:
    """
    Returns a boolean Series where True indicates the value
    contains only digits or punctuation (no alphabetic characters).
    """
    return series.astype(str).str.match(r'^[\d\W]+$')
```

```python
punctuation_query_rows = is_numeric_or_punctuation(data['QueryText'])
data[punctuation_query_rows]
```

```python
punctuation_ans_rows = is_numeric_or_punctuation(data['KccAns'])
data[punctuation_ans_rows]
```

```python
data = data[~punctuation_query_rows]
data = data[~punctuation_ans_rows]
data.shape
```

```python
unwanted_querytext_keywords = ["number", "call", "market", "weather"]
data[data['QueryText'].str.contains('|'.join(unwanted_querytext_keywords), case=False)]
```

```python
unwanted_kccans_keywords = ["given the details", "gave the details", "explained", "contact", "gave details", "given", "market"]
data[data['KccAns'].str.contains('|'.join(unwanted_kccans_keywords), case=False)]
```

```python
unwanted_querytext_keywords = ["weather", "mansoon", "chances of rain"]
unwanted_kccans_keywords = ["weather", "mansoon", "chances of rain"]

def is_unwanted_kccans(text):
    text_lower = text.lower()
    contains_keyword = any(word in text_lower for word in unwanted_kccans_keywords)
    if contains_keyword:
        return True
    return False
```

```python
data = data[~(
    (
        data["QueryText"].str.contains('|'.join(unwanted_querytext_keywords), na=False)
    ) |
    (
        data["KccAns"].apply(is_unwanted_kccans)
    )
)]
```
